```python
from imblearn.over_sampling import SMOTE
```

```python
smote=SMOTE(k_neighbors=5)
```

```python
x_res,y_res=smote.fit_resample(x,y)
```

```python
y_res.value_counts()
```

```python
sns.countplot(x=y_res)
```

```python
# feature selection
```

```python
from sklearn.feature_selection import SelectKBest,chi2
```

```python
skf=SelectKBest(k=5,score_func=chi2)
```

```python
x_new = skf.fit_transform(x_res,y_res)
```

```python
x_new.shape
```

```python
skf.get_support()
```

```python
skf.get_feature_names_out()
```

```python
x=x_res.loc[:,skf.get_support()]
```

```python
x
```

```python
x_res.shape
```

```python
y_res.shape
```

```python
x.describe()
```

```python
# cross validation
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y_res,random_state=0,
                                               test_size=0.2,
                                               stratify=y_res)
```

```python
x_train.shape,y_train.shape
```

```python
# feature scaling
```

```python
from sklearn.preprocessing import MinMaxScaler
sca=MinMaxScaler()

x_train_scaled=sca.fit_transform(x_train)
x_test_scaled=sca.transform(x_test)
```

```python
x_train_scaled.shape
```

```python
y_train.shape
```

```python
#build the model
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
params = {'random_state':[0,42],
          'criterion':['gini','entropy'],
          'max_features':['sqrt','log2'],
          'n_estimators':[20,30,100]}
```

```python
grid=GridSearchCV(estimator=RandomForestClassifier(),
                  param_grid=params,
                  scoring='accuracy',
                  cv=5)
                  
```

```python
grid.fit(x_train_scaled,y_train)
```

```python
y_pred = grid.predict(x_test_scaled)
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay,accuracy_score
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
